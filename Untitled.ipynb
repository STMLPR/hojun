# data processing(to International System Units)

```python
def convert_units(df):
    # 데이터프레임 복사
    df = df.copy()
    
    # 화씨를 섭씨로 변환 (°F = °C × 9/5 + 32)
    df['Temperature_C'] = (df['Body_Temperature(F)'] - 32) * 5/9
    
    # 키를 cm로 변환 (1피트 = 30.48cm, 1인치 = 2.54cm)
    df['Height_cm'] = (df['Height(Feet)'] * 30.48) + (df['Height(Remainder_Inches)'] * 2.54)
    
    # 체중을 kg으로 변환 (1파운드 = 0.453592kg)
    df['Weight_kg'] = df['Weight(lb)'] * 0.453592
    
    # BMI 계산 추가
    df['BMI'] = df['Weight_kg'] / ((df['Height_cm']/100) ** 2)
    
    # 기존 컬럼 삭제
    columns_to_drop = ['Body_Temperature(F)', 'Height(Feet)', 
                      'Height(Remainder_Inches)', 'Weight(lb)']
    df = df.drop(columns_to_drop, axis=1)
    
    return df

# 훈련 데이터와 테스트 데이터에 적용
train = convert_units(train)
test = convert_units(test)

# 변환된 데이터 확인
print("\n=== 단위 변환 후 데이터 범위 ===")
print("\n체온 범위:")
print(f"최소: {train['Temperature_C'].min():.1f}°C")
print(f"최대: {train['Temperature_C'].max():.1f}°C")

print("\n키 범위:")
print(f"최소: {train['Height_cm'].min():.1f}cm")
print(f"최대: {train['Height_cm'].max():.1f}cm")

print("\n체중 범위:")
print(f"최소: {train['Weight_kg'].min():.1f}kg")
print(f"최대: {train['Weight_kg'].max():.1f}kg")

print("\nBMI 범위:")
print(f"최소: {train['BMI'].min():.1f}")
print(f"최대: {train['BMI'].max():.1f}")

# 변환된 데이터의 처음 몇 행 확인
print("\n=== 변환된 데이터 샘플 ===")
print(train.head())
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# 1. 수치형 변수들의 왜도와 첨도 확인
numeric_features = ['Exercise_Duration', 'Temperature_C', 'BPM', 
                   'Weight_kg', 'Height_cm', 'Age', 'BMI']

# 왜도와 첨도 계산
skew_kurt = pd.DataFrame({
    'Skewness': train[numeric_features].skew(),
    'Kurtosis': train[numeric_features].kurtosis()
}).round(3)

print("왜도와 첨도:")
print(skew_kurt.sort_values(['Kurtosis'], ascending=False))

# 로그 변환이 필요한 변수 선정 
# |왜도| > 1 또는 |첨도| > 3
log_features = skew_kurt[
    (abs(skew_kurt['Skewness']) > 1) | 
    (abs(skew_kurt['Kurtosis']) > 3)
].index.tolist()

print("\n로그 변환이 필요한 변수들:", log_features)

# 만약 로그 변환이 필요한 변수가 있다면, 변환 전후 비교
if log_features:
    for feature in log_features:
        plt.figure(figsize=(12, 4))
        
        plt.subplot(1, 2, 1)
        sns.histplot(data=train, x=feature, kde=True)
        plt.title(f'Original {feature}\nSkew: {skew_kurt.loc[feature, "Skewness"]:.3f}, Kurt: {skew_kurt.loc[feature, "Kurtosis"]:.3f}')
        
        plt.subplot(1, 2, 2)
        transformed_data = np.log1p(train[feature])
        sns.histplot(x=transformed_data, kde=True)
        plt.title(f'Log Transformed {feature}\nSkew: {transformed_data.skew():.3f}, Kurt: {transformed_data.kurtosis():.3f}')
        
        plt.tight_layout()
        plt.show()

# 4. 로그 변환과 스케일링 적용
from sklearn.preprocessing import StandardScaler

# 로그 변환
X_transformed = train[numeric_features].copy()
for feature in log_features:
    X_transformed[feature] = np.log1p(X_transformed[feature])

# 스케일링
scaler = StandardScaler()
X_scaled = pd.DataFrame(
    scaler.fit_transform(X_transformed),
    columns=X_transformed.columns,
    index=X_transformed.index
)

# 최종 결과 확인
print("\n변환 후 왜도:")
print(X_scaled.skew().sort_values(ascending=False))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline

# 특성 구분
numeric_features = ['Exercise_Duration', 'Temperature_C', 'BPM', 
                   'Height_cm', 'Weight_kg', 'Age', 'BMI']
categorical_features = ['Weight_Status', 'Gender']

# 전처리 파이프라인 정의
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(drop='first', sparse_output=False), categorical_features)
    ])

# X, y 분리
X = train[numeric_features + categorical_features]
y = train['Calories_Burned']

# 전처리 적용
X_transformed = preprocessor.fit_transform(X)

# 변환된 특성 이름 생성
numeric_names = numeric_features
categorical_names = (preprocessor
                    .named_transformers_['cat']
                    .get_feature_names_out(categorical_features))
feature_names = numeric_names + list(categorical_names)

# 결과를 데이터프레임으로 변환
X_processed = pd.DataFrame(
    X_transformed, 
    columns=feature_names,
    index=X.index
)
```
