# data processing(to International System Units)

```python
def convert_units(df):
    # 데이터프레임 복사
    df = df.copy()
    
    # 화씨를 섭씨로 변환 (°F = °C × 9/5 + 32)
    df['Temperature_C'] = (df['Body_Temperature(F)'] - 32) * 5/9
    
    # 키를 cm로 변환 (1피트 = 30.48cm, 1인치 = 2.54cm)
    df['Height_cm'] = (df['Height(Feet)'] * 30.48) + (df['Height(Remainder_Inches)'] * 2.54)
    
    # 체중을 kg으로 변환 (1파운드 = 0.453592kg)
    df['Weight_kg'] = df['Weight(lb)'] * 0.453592
    
    # BMI 계산 추가
    df['BMI'] = df['Weight_kg'] / ((df['Height_cm']/100) ** 2)
    
    # 기존 컬럼 삭제
    columns_to_drop = ['Body_Temperature(F)', 'Height(Feet)', 
                      'Height(Remainder_Inches)', 'Weight(lb)']
    df = df.drop(columns_to_drop, axis=1)
    
    return df

# 훈련 데이터와 테스트 데이터에 적용
train = convert_units(train)
test = convert_units(test)

# 변환된 데이터 확인
print("\n=== 단위 변환 후 데이터 범위 ===")
print("\n체온 범위:")
print(f"최소: {train['Temperature_C'].min():.1f}°C")
print(f"최대: {train['Temperature_C'].max():.1f}°C")

print("\n키 범위:")
print(f"최소: {train['Height_cm'].min():.1f}cm")
print(f"최대: {train['Height_cm'].max():.1f}cm")

print("\n체중 범위:")
print(f"최소: {train['Weight_kg'].min():.1f}kg")
print(f"최대: {train['Weight_kg'].max():.1f}kg")

print("\nBMI 범위:")
print(f"최소: {train['BMI'].min():.1f}")
print(f"최대: {train['BMI'].max():.1f}")

# 변환된 데이터의 처음 몇 행 확인
print("\n=== 변환된 데이터 샘플 ===")
print(train.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# seaborn 스타일 설정
sns.set_theme()  # 새로운 seaborn 스타일 적용 방식
plt.rcParams['font.family'] = 'Malgun Gothic'  # 한글 폰트 설정

# 그래프 크기 설정
plt.figure(figsize=(15, 10))

# 1. 수치형 변수들의 분포 및 상관관계
sns.pairplot(data=train[['Exercise_Duration', 'Temperature_C', 'BPM', 
                        'Weight_kg', 'Height_cm', 'Age', 'Calories_Burned']], 
             diag_kind='kde')
plt.suptitle('주요 변수들의 산점도 매트릭스', y=1.02, fontsize=16)

# 2. 히트맵으로 상관관계 시각화
plt.figure(figsize=(10, 8))
correlation = train[['Exercise_Duration', 'Temperature_C', 'BPM', 
                    'Weight_kg', 'Height_cm', 'Age', 'Calories_Burned']].corr()
sns.heatmap(correlation, annot=True, cmap='RdYlBu_r', center=0)
plt.title('변수간 상관관계 히트맵', pad=20, fontsize=16)

# 3. BMI와 Weight_Status의 관계
plt.figure(figsize=(10, 6))
sns.boxplot(x='Weight_Status', y='BMI', data=train)
plt.title('체중 상태별 BMI 분포', pad=20, fontsize=16)

# 4. 성별에 따른 칼로리 소모량
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Calories_Burned', data=train)
plt.title('성별에 따른 칼로리 소모량 분포', pad=20, fontsize=16)

# 5. 운동 시간과 칼로리 소모량의 관계 (체중 상태별)
plt.figure(figsize=(12, 6))
sns.scatterplot(data=train, x='Exercise_Duration', y='Calories_Burned', 
                hue='Weight_Status', alpha=0.6)
plt.title('운동 시간과 칼로리 소모량의 관계 (체중 상태별)', pad=20, fontsize=16)

plt.tight_layout()
plt.show()
```
